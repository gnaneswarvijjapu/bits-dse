```python
no=int(input("enter the no of nodes /n"))
```

```python
#creation of adjacency matrix
mat=[[0 for i in range(no)] for j in range(no)]
m=[]
for i in range(no):
    m.append(input("enter {}th node name".format(i+1)))

for k in range(no):
    mat[k][k]=1
print("Enter the edges seperate by coma one by one ex: A,B")
while(True):
    inp=input()
    if inp:
        k=inp.split(",")
        x=int(m.index(k[0]))
        y=int(m.index(k[1]))
        mat[x][y]=1
        mat[y][x]=1
        
    else:
        break

```

```python
to_plot=[]
nodes=['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O']
for i in range(0,len(mat)):
    sum=0
    for j in range(0,len(mat)):
        if mat[i][j]==1:
            sum=sum+1
    to_plot.append(sum-1)
            
```
